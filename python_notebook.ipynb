## Lets investigate what is $e^{ix}$ ##

There is a power series expansion for $e^{ix}$

$e^{x} = 1 + \frac{x}{1!} +  \frac{x^2}{2!}+ \frac{x^3}{3!}$

$e^{ix} = 1 + \frac{ix}{1!} -  \frac{x^2}{2!} - \frac{ix^3}{3!} +  \frac{x^4}{4!} + \frac{ix^5}{5!} - \frac{x^6}{6!}+ \dots$

```python
import math

def calculate_e_power_ix(x_val,number_of_powers=10):
    answer = 0
    for i in range(number_of_powers):
        answer+=pow((0+1j)*x_val,i)/math.factorial(i)
    return answer
```

```python
ans=calculate_e_power_ix(7)
print("example")
print(ans)
```

```python
print(calculate_e_power_ix(10*math.pi))
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Define an array of complex numbers (using 'j' for the imaginary unit)
complex_numbers = []
for i in np.arange (0.0,10*math.pi,0.9):
    complex_numbers.append(calculate_e_power_ix(i))
print("complex numbers")
print(complex_numbers)
#complex_numbers = [2 + 3j, -3 - 2j, -1 + 4j, 3 - 3j]

# Extract real parts for X-axis and imaginary parts for Y-axis
X = [z.real for z in complex_numbers]
Y = [z.imag for z in complex_numbers]

# Initialize the plot
fig, ax = plt.subplots(figsize=(6, 6))

# Draw the point markers
ax.scatter(X, Y, color='red', s=3, zorder=5)

# Add text labels next to each plotted point
#for z in complex_numbers:
    #ax.text(z.real + 0.15, z.imag + 0.15, f'{z.real} + {z.imag}j', fontsize=10)

# Configure the complex plane cross-axes style
ax.axhline(0, color='black', linewidth=1.2)  # Real axis
ax.axvline(0, color='black', linewidth=1.2)  # Imaginary axis

# Customizing grids and labels
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_xlabel('Real Axis (Re)', fontsize=12)
ax.set_ylabel('Imaginary Axis (Im)', fontsize=12)
ax.set_title('Complex Plane Plot', fontsize=14)

# Keep the scale proportional so circles look like circles
ax.set_aspect('equal')

# Set layout padding and display
plt.xlim(-20, 20)
plt.ylim(-20, 20)
plt.tight_layout()
plt.show()
```
